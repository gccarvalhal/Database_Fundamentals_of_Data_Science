```python
#importar
import psycopg2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

```python
#conect to de database
connection = psycopg2.connect(
    database = "happiness",
    user = "postgres",
    password = "admin",
    host = "localhost",
    port = 5432,
    options = "-c search_path=happiness"
)
```

```python
df_extra = pd.read_sql('SELECT * FROM Rank_CPLP',connection)
df_extra.head()
```

```python
#Create plot
fig = plt.figure(figsize=(25,10))
sns.set_style("ticks")
plot = sns.lineplot(data=df_extra, x='year', y='rank', hue='country', linewidth = 2,  markers=True, marker='o', markersize=10)
plot.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=15)
plot.invert_yaxis()
plt.xticks(range(2015, 2020, 1), size='large' )
plt.yticks(size='large' )
plt.xlabel('Year',fontsize=20)
plt.ylabel('Rank',fontsize=20)
plt.title('CPLP happiness per year', fontsize=30)

#Save figure
plt.savefig('CPLP happiness per year.png', dpi=300)

plt.show()
```
